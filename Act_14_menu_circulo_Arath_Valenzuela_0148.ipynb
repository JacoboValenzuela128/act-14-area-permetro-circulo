Este programa tiene como objetivo aplicar conceptos básicos de geometría euclidiana mediante programación estructurada en Python, un enfoque ideal para estudiantes de preparatoria que están iniciando en el desarrollo de software. El código está diseñado de forma modular y procedural, lo que significa que cada tarea se resuelve en funciones independientes sin recurrir a clases ni objetos.

Donde r representa el radio de la figura y π es la constante matemática aproximada a $3.14159...$

* El flujo del programa se controla mediante una función $menu()$ que muestra las opciones al usuario y se mantiene en ejecución gracias a un ciclo while.
* Dentro del ciclo, las estructuras condicionales if, elif y else determinan qué función llamar según la elección del usuario.
* Cada función ($areacirculo()$, $perimetrocirculo()$ y $volumenesfera()$) solicita el valor del radio en el momento de ejecutarse, realiza el cálculo correspondiente y muestra el resultado en pantalla.
* No se incluye validación de entrada para mantener el código sencillo y enfocado en la lógica básica de programación.
* La interfaz se ha decorado con emojis para hacer la interacción más amigable y visual.
* Al seleccionar la opción 4, el ciclo termina y el programa finaliza de manera ordenada.

Para realizar los cálculos geométricos se utilizan las siguientes fórmulas matemáticas fundamentales:
1. Área del círculo:
* $A = π * r^2$
2. Perímetro (circunferencia) del círculo:
* $P = 2 * π * r$
3. Volumen de la esfera:
* $V = 3/4 * r^3$

```python
import math

def areacirculo():
    """Calcula y muestra el área de un círculo."""
    print("\n📐 Cálculo del Área del Círculo")
    radio = float(input("📥 Ingresa el radio del círculo: "))
    area = math.pi * (radio ** 2)
    print(f"✅ El área del círculo con radio {radio} es: {area:.2f} unidades²")

def perimetrocirculo():
    """Calcula y muestra el perímetro de un círculo."""
    print("\n📏 Cálculo del Perímetro del Círculo")
    radio = float(input("📥 Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"✅ El perímetro del círculo con radio {radio} es: {perimetro:.2f} unidades")

def volumenesfera():
    """Calcula y muestra el volumen de una esfera."""
    print("\n🌐 Cálculo del Volumen de la Esfera")
    radio = float(input("📥 Ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * (radio ** 3)
    print(f"✅ El volumen de la esfera con radio {radio} es: {volumen:.2f} unidades³")

def menu():
    """Muestra el menú interactivo y controla el flujo del programa."""
    opcion = 0
    while opcion != 4:
        print("\n🔹 ══════ MENÚ PRINCIPAL ══════ 🔹")
        print("1️⃣  Calcular área del círculo")
        print("2️⃣  Calcular perímetro del círculo")
        print("3️⃣  Calcular volumen de la esfera")
        print("4️⃣  Salir del programa")
        print("═══════════════════════════════")

        opcion = int(input("👉 Elige una opción (1-4): "))

        if opcion == 1:
            areacirculo()
        elif opcion == 2:
            perimetrocirculo()
        elif opcion == 3:
            volumenesfera()
        elif opcion == 4:
            print("\n👋 ¡Gracias por usar el programa! Hasta pronto. ✨")
        else:
            print("⚠️ Opción no válida. Por favor, selecciona un número del 1 al 4.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
